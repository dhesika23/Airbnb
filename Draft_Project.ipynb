```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
df = pd.read_csv('/content/drive/MyDrive/EDA/Airbnb NYC 2019.csv')
```

```python
# Check for missing values
missing_values = df.isna().sum()

print(missing_values)
```

```python
new_df = df[['id','name','host_id','host_name','neighbourhood_group','neighbourhood','room_type','price','minimum_nights',
             'number_of_reviews','calculated_host_listings_count','availability_365']]
```

```python
hosts_data = {
    'host_id': [2787, 2845, 4632, 4869, 7192, ...],  # Add all host_id values
    'host_name': ['John', 'Jennifer', 'Elisabeth', 'LisaRoxanne', 'Laura', ...]  # Add all host_name values
}

hosts_df = pd.DataFrame(hosts_data)

# Merge the main DataFrame with the hosts DataFrame based on 'host_id'
df_with_host_name = df.merge(hosts_df, on='host_id', how='left')
df_with_host_name
```

```python
aggregated_data = df.groupby('neighbourhood_group')['calculated_host_listings_count'].max().reset_index()
aggregated_data.sort_values(by='calculated_host_listings_count', ascending=True).head(5)
```
