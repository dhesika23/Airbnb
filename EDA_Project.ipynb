```python
# Check Unique Values for each variable.
unique_values = {}
for column in df.columns:
    unique_values[column] = df[column].unique()

# Print unique values for each variable
for column, values in unique_values.items():
    print(f"Unique values for '{column}':\n{values}\n")
```

```python
# Create a copy of the original DataFrame 'df' and assign it to 'df_copy'
df_copy=df.copy()

# Remove the specified columns from 'df_copy'
df_copy.drop(['reviews_per_month','last_review'],axis=1,inplace=True)

# Identify row indices with missing values and store them
host_name_null_values=list(df_copy[pd.isnull(df_copy["host_name"])].index.values)
df_copy.iloc[host_name_null_values,[3]] ='Unknown'

#Drop the 'name' column from the 'df_copy' as it has no valuable info
df_copy.drop(['name'],axis=1,inplace=True)

# Replace missing values in the 'host_name' column
null_values_in_data=df_copy.isnull().sum().sort_values(ascending=False)

# Calculate the mean value of the 'price' column
zero_price_rows = df_copy[df_copy['price'] == 0]
mean_value_of_price = df_copy['price'].mean()

print("Mean Price:", mean_value_of_price)
print("New columns", df_copy.columns)
print("Null Host Name",df_copy['host_name'].isnull().sum())
print(null_values_in_data)
```

```python
#let's check the shape again
df_copy.shape
```

```python
# Chart - 1 visualization code
# Set the figure size
plt.figure(figsize=(8, 7))

# Create the scatterplot with customizations
sns.scatterplot(
    x=df_copy.longitude,
    y=df_copy.latitude,
    hue=df_copy.neighbourhood_group,
    palette='Set1',  #change the color palette here
    s=45,  # Adjust the size of the markers
    alpha=0.6,  # Adjust the marker transparency
    edgecolor='k',  # Add a black edge to markers
)

# Add title and labels
plt.title('Latitude and Longitude in range of Neighbourhood')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# Add a legend with custom title
plt.legend(title='Neighbourhood Group')

# Show the plot
plt.show()
```

```python
# Chart - 2 visualization code
sns.boxplot(x='availability_365', data=df_copy)
plt.xlabel('Availability (in days)')
plt.title('Distribution of Availability')
plt.show()
```

```python
# Chart - 3 visualization code
area_reviews = df_copy.groupby(['neighbourhood_group'])['number_of_reviews'].max().reset_index()
area_reviews
area = area_reviews['neighbourhood_group']
review = area_reviews['number_of_reviews']

fig = plt.figure(figsize=(10, 6))

# Creating the bar plot
plt.bar(area, review, color='green', width=0.2)

plt.xlabel("Neighbourhood Group")
plt.ylabel("Number of Reviews")
plt.title("Number of Reviews by Neighbourhood Group")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Chart - 4 visualization code
top_host_names = df_copy['host_name'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_host_names.index, y=top_host_names.values)
plt.xlabel('Host Name')
plt.ylabel('Count')
plt.title('Top Host Names by Number of Listings')
plt.xticks(rotation=45)
plt.show()
```

```python
# Chart - 5 visualization code
# Group by 'neighbourhood' and calculate the sum of reviews for each neighbourhood
neighbourhood_reviews = df_copy.groupby('neighbourhood')['number_of_reviews'].sum().reset_index()

# Sort by the sum of reviews in descending order and select the top N neighbourhoods
top_neighbourhoods = neighbourhood_reviews.sort_values(by='number_of_reviews', ascending=False).head()

plt.figure(figsize=(12, 8))
sns.barplot(x='number_of_reviews', y='neighbourhood', data=top_neighbourhoods, orient='h')
plt.xlabel('Number of Reviews')
plt.ylabel('Neighbourhood')
plt.title('Top Neighbourhoods by Number of Reviews')

# Rotate labels to 0 degrees for better readability
plt.xticks(rotation=0)

plt.tight_layout()
plt.show()
```

```python
# Chart - 6 visualization code
sns.boxplot(x='minimum_nights', data=df_copy)
plt.xlabel('Minimum Nights')
plt.title('Distribution of Minimum Nights')
plt.show()
```

```python
# Chart - 7 visualization code
plt.hist(df_copy['price'], bins=20)
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Prices')
plt.show()
```

```python
# Chart - 8 visualization code
# Calculate the room type counts
room_type_counts = df_copy['room_type'].value_counts()
# Create a pie chart
plt.figure(figsize=(7, 8))
plt.pie(room_type_counts, labels=room_type_counts.index, colors=['yellow', 'black', 'violet'])
plt.title('Room Type Distribution', fontsize=20)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Correlation Heatmap visualization code
corr_matrix = df_copy.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Pair Plot visualization code
# Select numerical columns for the pair plot
numerical_columns = ['price', 'minimum_nights', 'number_of_reviews', 'calculated_host_listings_count', 'availability_365']

# Create a pair plot
sns.pairplot(df_copy[numerical_columns])
plt.show()
```
